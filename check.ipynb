```python
import numpy as np
import matplotlib.pyplot as plt

from keras.preprocessing import image
def prepare(img_path):
    img = load_img(img_path, target_size=(224,224))
    x = img_to_array(img)
    x = x/255
    return np.expand_dims(x, axis=0)

```

```python

result = model.predict([prepare(path)])
```
